```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Función para actualizar la gráfica
def update_plot(frequency):
    # Señal 1: Onda sinusoidal
    t = np.linspace(0, 1, 100, endpoint=False)
    signal1 = np.sin(2 * np.pi * frequency * t)

    # Señal 2: Onda cuadrada
    signal2 = np.sign(np.sin(2 * np.pi * frequency/2 * t)) #frecuencia de la onda cuadrada la mitad de la sinusoidal

    # Realizar la convolución
    convolved_signal = np.convolve(signal1, signal2, mode='same')

    # Actualizar la gráfica
    plt.clf() # Limpia la figura anterior

    plt.figure(figsize=(12, 8))
    plt.subplot(3, 1, 1)
    plt.plot(t, signal1)
    plt.title(f'Señal 1: Onda sinusoidal de {frequency} Hz')
    plt.xlabel('Tiempo [s]')
    plt.ylabel('Amplitud')

    plt.subplot(3, 1, 2)
    plt.plot(t, signal2)
    plt.title(f'Señal 2: Onda cuadrada de {frequency/2} Hz')
    plt.xlabel('Tiempo [s]')
    plt.ylabel('Amplitud')

    plt.subplot(3, 1, 3)
    plt.plot(t, convolved_signal)
    plt.title('Convolución de las dos señales')
    plt.xlabel('Tiempo [s]')
    plt.ylabel('Amplitud')

    plt.tight_layout()
    plt.show()


# Crear el slider de frecuencia
frequency_slider = widgets.FloatSlider(
    value=5,  # Frecuencia inicial
    min=1,
    max=20,
    step=0.5,
    description='Frecuencia (Hz):'
)

# Interactuar el slider con la función de actualización
widgets.interactive(update_plot, frequency=frequency_slider)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Función para actualizar la gráfica
def update_plot(frequency):
    # Definir ejes de tiempo:
    # Señal 1: baja resolución (100 puntos)
    t_low = np.linspace(0, 1, 100, endpoint=False)
    # Señal 2: alta resolución (1000 puntos)
    t_high = np.linspace(0, 1, 1000, endpoint=False)

    # Señal 1: Onda sinusoidal
    signal1 = np.sin(2 * np.pi * frequency * t_low)

    # Señal 2: Onda cosenoidal modificada (alta FS)
    # Se aumenta la amplitud, se añade un desfase y un offset para diferenciarla
    signal2 = 1.5 * np.cos(2 * np.pi * frequency * t_high + np.pi/4) - 0.5

    # Convolución de ambas señales
    # La convolución se realiza entre arrays de distinto tamaño.
    # Con mode='same' se obtiene un resultado de longitud igual al mayor (en este caso, 1000 puntos)
    convolved_signal = np.convolve(signal1, signal2, mode='same')

    # Generar un eje de tiempo para la señal resultante
    t_conv = np.linspace(0, 1, len(convolved_signal), endpoint=False)

    # Actualizar la gráfica
    plt.clf()  # Limpia la figura anterior
    plt.figure(figsize=(12, 10))

    plt.subplot(3, 1, 1)
    plt.plot(t_low, signal1, marker='o', linestyle='-')
    plt.title(f'Señal 1: Onda sinusoidal de {frequency} Hz\n(Baja FS: 100 puntos en 1 s)')
    plt.xlabel('Tiempo [s]')
    plt.ylabel('Amplitud')

    plt.subplot(3, 1, 2)
    plt.plot(t_high, signal2, 'g-')
    plt.title(f'Señal 2: Onda cosenoidal modificada de {frequency} Hz\n(Alta FS: 1000 puntos en 1 s)')
    plt.xlabel('Tiempo [s]')
    plt.ylabel('Amplitud')

    plt.subplot(3, 1, 3)
    plt.plot(t_conv, convolved_signal, 'm-')
    plt.title('Convolución de las dos señales')
    plt.xlabel('Tiempo [s]')
    plt.ylabel('Amplitud')

    plt.tight_layout()
    plt.show()

# Crear el slider de frecuencia
frequency_slider = widgets.FloatSlider(
    value=5,      # Frecuencia inicial
    min=1,
    max=20,
    step=0.5,
    description='Frecuencia (Hz):'
)

# Interactuar el slider con la función de actualización
widgets.interactive(update_plot, frequency=frequency_slider)
```
